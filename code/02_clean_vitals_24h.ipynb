### Vitals Extraction Pipeline (ITEMID Mapping → 24h Aggregation)

```python
import os, json
import pandas as pd

DATA_DIR = "/content/drive/MyDrive/mimic_data"
OUT_DIR  = "/content/drive/MyDrive/mimic_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

D_ITEMS_PATH = os.path.join(DATA_DIR, "D_ITEMS.csv")
```

```python
# Build Vitals ITEMID Mapping (D_ITEMS → vital_itemids.json)
VITAL_PATTERNS = {
    "heart_rate": ["heart rate"],
    "resp_rate": ["respiratory rate"],
    "spo2": ["spo2"],
    "temperature": ["temperature"],
    "map": ["mean bp", "mean arterial pressure", "mean arterial"],
}

d = pd.read_csv(D_ITEMS_PATH, usecols=["ITEMID", "LABEL"])
d.columns = d.columns.str.lower()
d["label_l"] = d["label"].astype(str).str.lower()

mapping = {}
for concept, pats in VITAL_PATTERNS.items():
    mask = False
    for p in pats:
        mask = mask | d["label_l"].str.contains(p, na=False)
    itemids = sorted(d.loc[mask, "itemid"].dropna().astype(int).unique().tolist())
    mapping[concept] = itemids

out_json = os.path.join(OUT_DIR, "vital_itemids.json")
with open(out_json, "w") as f:
    json.dump(mapping, f, indent=2)

print("✅ Saved:", out_json)
for k, v in mapping.items():
    print(k, "n=", len(v), "sample=", v[:10])

# export an audit table to manually review/remove irrelevant labels
# (temperature often matches many itemids and benefits from manual filtering)
audit = []
for concept, itemids in mapping.items():
    sub = d[d["itemid"].isin(itemids)][["itemid","label"]].drop_duplicates()
    sub["concept"] = concept
    audit.append(sub)
audit_df = pd.concat(audit).sort_values(["concept","itemid"])
audit_csv = os.path.join(OUT_DIR, "vital_itemid_audit.csv")
audit_df.to_csv(audit_csv, index=False)
print("✅ Saved audit:", audit_csv)
```

```python
D_ITEMS_PATH = os.path.join(DATA_DIR, "D_ITEMS.csv")

VITAL_PATTERNS = {
    "heart_rate": ["heart rate"],
    "resp_rate": ["respiratory rate"],
    "spo2": ["spo2"],
    "temperature": ["temperature"],
    "map": [
        "mean arterial pressure",
        r"\bmap\b",
        "arterial bp.*mean",
        "mean bp",
        "bp.*mean",
    ],
}

d = pd.read_csv(D_ITEMS_PATH, usecols=["ITEMID", "LABEL"])
d.columns = d.columns.str.lower()
d["label_l"] = d["label"].astype(str).str.lower()

mapping = {}
for concept, pats in VITAL_PATTERNS.items():
    mask = False
    for p in pats:
        mask = mask | d["label_l"].str.contains(p, na=False, regex=True)
    itemids = sorted(d.loc[mask, "itemid"].dropna().astype(int).unique().tolist())
    mapping[concept] = itemids

out_json = os.path.join(OUT_DIR, "vital_itemids.json")
with open(out_json, "w") as f:
    json.dump(mapping, f, indent=2)

print("✅ Saved:", out_json)
for k, v in mapping.items():
    print(k, "n=", len(v), "sample=", v[:20])

# audit
audit = []
for concept, itemids in mapping.items():
    sub = d[d["itemid"].isin(itemids)][["itemid","label"]].drop_duplicates()
    sub["concept"] = concept
    audit.append(sub)
audit_df = pd.concat(audit).sort_values(["concept","itemid"])
audit_csv = os.path.join(OUT_DIR, "vital_itemid_audit.csv")
audit_df.to_csv(audit_csv, index=False)
print("✅ Saved audit:", audit_csv)
```

```python
# Aggregate 24h Vitals (CHARTEVENTS → vitals_24h.parquet)
COHORT_PATH = os.path.join(OUT_DIR, "cohort_icustay.parquet")
ITEMID_PATH = os.path.join(OUT_DIR, "vital_itemids.json")
SUBSET_PATH = os.path.join(DATA_DIR, "chartevents_vitals_subset.parquet")

# 1) cohort
cohort = pd.read_parquet(COHORT_PATH)[["icustay_id", "intime"]].copy()
cohort["intime"] = pd.to_datetime(cohort["intime"], errors="coerce")
intime_map = cohort.set_index("icustay_id")["intime"]

# 2) itemid -> concept
with open(ITEMID_PATH, "r") as f:
    mapping = json.load(f)
itemid_to_concept = {int(iid): concept for concept, ids in mapping.items() for iid in ids}

# 3) load subset
df = pd.read_parquet(SUBSET_PATH)
df["charttime"] = pd.to_datetime(df["charttime"], errors="coerce")
df = df.dropna(subset=["icustay_id", "charttime", "itemid", "valuenum"])

# 4) 24h filter
df["intime"] = df["icustay_id"].astype(int).map(intime_map)
df = df.dropna(subset=["intime"])
df = df[(df["charttime"] >= df["intime"]) & (df["charttime"] < df["intime"] + pd.Timedelta(hours=24))]

# 5) concept
df["itemid"] = df["itemid"].astype(int)
df["concept"] = df["itemid"].map(itemid_to_concept)

# 6) temperature unify to Celsius
TEMP_C_ITEMIDS = {676, 223762}
TEMP_F_ITEMIDS = {678, 223761}
is_temp = df["concept"] == "temperature"
mask_f = is_temp & df["itemid"].isin(TEMP_F_ITEMIDS)
df.loc[mask_f, "valuenum"] = (df.loc[mask_f, "valuenum"] - 32.0) * (5.0 / 9.0)
mask_temp = is_temp & df["itemid"].isin(TEMP_C_ITEMIDS | TEMP_F_ITEMIDS)
df.loc[mask_temp, "concept"] = "temperature_c"

# 7) aggregate
df = df.sort_values(["icustay_id", "concept", "charttime"])
g = df.groupby(["icustay_id", "concept"])["valuenum"]
agg = g.agg(["mean","min","max","std","median","count"]).reset_index()
last = df.groupby(["icustay_id","concept"]).tail(1)[["icustay_id","concept","valuenum"]].rename(columns={"valuenum":"last"})
agg = agg.merge(last, on=["icustay_id","concept"], how="left")

wide = agg.pivot(index="icustay_id", columns="concept")
wide.columns = [f"{concept}_{stat}" for stat, concept in wide.columns]
wide = wide.reset_index()

out_path = os.path.join(OUT_DIR, "vitals_24h.parquet")
wide.to_parquet(out_path, index=False, compression="snappy")
print("✅ saved:", out_path, "shape:", wide.shape)
```

```python
# merge cohort + vitals → train_structured_vitals_24h.parquet
cohort = pd.read_parquet("/content/drive/MyDrive/mimic_outputs/cohort_icustay.parquet")
vitals = pd.read_parquet("/content/drive/MyDrive/mimic_outputs/vitals_24h.parquet")

train = cohort.merge(vitals, on="icustay_id", how="left")
train.to_parquet("/content/drive/MyDrive/mimic_outputs/train_structured_vitals_24h.parquet",
                 index=False, compression="snappy")
print(train.shape)
```
